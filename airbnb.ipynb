**Copy data dari dataset asli**

```python
#membuat data baru
dfClean = df.dropna(subset='name')
```

```python
#menghapus kolom yang tidak di gunakan
dfClean.drop(columns=['number_of_reviews_ltm;','Unnamed: 0','id','last_review','reviews_per_month','host_id'],axis=1,inplace=True)
```

```python
# Cek apakah ada nilai yang hilang
print(dfClean.isnull().sum())
```

```python
#Mengecek Data yang duplikat
dfClean.duplicated().sum()

#Terdapat data duplikat, hapus data duplikat
dfClean.drop_duplicates(keep='first', inplace=True)
```

```python
#Mengecek kembali mengenai missing value dan tipe data
#Menampilkan jumlah baris dan kolom dari dataset
print(f'Jumlah baris dan kolom: {dfClean.shape}')

#Memberikan penilaian awal terhadap dataset
pd.DataFrame({
    'Type': dfClean.dtypes,
    'N/A (%)': dfClean.isna().mean() * 100,
    'Unique': dfClean.nunique(),
    'Sample': [dfClean[col].unique() for col in dfClean.columns]
})
```

```python
#Mengecek Data yang duplikat
dfClean.duplicated().sum()
```

```python
# Convert 'minimum_nights' to numeric, coercing errors to NaN
dfClean['minimum_nights'] = pd.to_numeric(dfClean['minimum_nights'], errors='coerce')
```

```python
#Mengecek data yang masi memiliki missing value
filtered_df = dfClean[
    dfClean['host_name'].isnull() |
    dfClean['availability_365'].isnull() 
]
filtered_df
```

```python
# Mengganti nilai NaN pada kolom 'host_name' dengan string kosong atau nilai default
dfClean['host_name'] = dfClean['host_name'].fillna('lou')

# Mengganti nilai NaN pada kolom 'number_of_reviews' dengan rata-rata
dfClean['number_of_reviews'] = pd.to_numeric(dfClean['number_of_reviews'], errors='coerce')
dfClean['number_of_reviews'] = dfClean['number_of_reviews'].fillna(dfClean['number_of_reviews'].mean())

# Mengganti nilai NaN pada kolom 'minimum_nights' dengan rata-rata
dfClean['minimum_nights'] = pd.to_numeric(dfClean['minimum_nights'], errors='coerce')
dfClean['minimum_nights'] = dfClean['minimum_nights'].fillna(dfClean['minimum_nights'].mean())
```

```python
# Mengganti nilai yang tidak dapat diubah langsung (seperti '19"') menjadi nilai yang sesuai
dfClean['availability_365'] = dfClean['availability_365'].replace({'"': ''}, regex=True).astype(float)
```

```python
# Mengganti nilai NaN pada kolom 'availability_365' dengan median
dfClean['availability_365'].fillna(dfClean['availability_365'].median(), inplace=True)
```

```python
# Mengganti nilai NaN pada kolom 'latitude' dan 'longitude' dengan median
dfClean['latitude'] = pd.to_numeric(dfClean['latitude'], errors='coerce')
dfClean['latitude'] = dfClean['latitude'].fillna(dfClean['latitude'].median())

dfClean['longitude'] = pd.to_numeric(dfClean['longitude'], errors='coerce')
dfClean['longitude'] = dfClean['longitude'].fillna(dfClean['longitude'].median())
```

```python
#Mengecek ulang data yang masi memiliki missing value
filtered_df = dfClean[
    dfClean['host_name'].isnull() |
    dfClean['availability_365'].isnull() |
    dfClean['minimum_nights'].isnull() 
]
filtered_df
```

**Mengganti nilai unik dengan dictionary**

```python
# Mengganti nilai '1500', '1', '1339' di kolom room_type
replace_dict = {
    '1500': 'Entire home/apt',
    '1': 'Private room',
    '1339': 'Shared room'
}

dfClean['room_type'] = dfClean['room_type'].replace(replace_dict)
```

```python
#Mengganti nilai tersebut di kolom minimum nights
dfClean['minimum_nights'] = dfClean['minimum_nights'].replace([3.000e+00, 6.000e+01, 7.000e+00, 1.000e+00, 2.500e+02, 2.000e+00, 1.500e+01], [3, 61, 7, 1, 250, 2, 15])
```

```python
# Mengganti nilai '100.5486', '13.7085', '13.7505' di kolom neighbourhood
replace_dict_neighbourhood = {
    '100.5486': 'Phra Pradaeng',
    '13.7085': 'Samut Sakhon',
    '13.7505': 'Nonthaburi'
}

dfClean['neighbourhood'] = dfClean['neighbourhood'].replace(replace_dict_neighbourhood)
```

```python
# Mengisi nilai nol dengan median berdasarkan room_type
dfClean['number_of_reviews'] = dfClean.groupby('room_type')['number_of_reviews'].transform(lambda x: x.replace(0, x.median()))
```

```python
#Mengecek kembali mengenai missing value dan tipe data
#Menampilkan jumlah baris dan kolom dari dataset
print(f'Jumlah baris dan kolom: {dfClean.shape}')

#Memberikan penilaian awal terhadap dataset
pd.DataFrame({
    'Type': dfClean.dtypes,
    'N/A (%)': dfClean.isna().mean() * 100,
    'Unique': dfClean.nunique(),
    'Sample': [dfClean[col].unique() for col in dfClean.columns]
})
```

```python
dfClean.isnull().sum()
```

**2.Apakah wilayah mempengaruhi harga sewa tipe ruangan di air bnb?**

```python
# Calculate price_per_night if not already done
dfClean['price_per_night'] = dfClean['price'] / dfClean['minimum_nights']

# Group by neighborhood and room type, and calculate the average price per night
neighborhood_room_type_avg_price = dfClean.groupby(['neighbourhood', 'room_type'])['price_per_night'].mean().reset_index()

# Find the average price per night for each neighborhood
neighborhood_avg_price = dfClean.groupby('neighbourhood')['price_per_night'].mean().reset_index()

# Sort neighborhoods by average price per night and select top 5 most expensive
top_5_expensive_neighborhoods = neighborhood_avg_price.sort_values(by='price_per_night', ascending=False).head(5)

# Find the room types in these top 5 neighborhoods
top_5_expensive_details = neighborhood_room_type_avg_price[neighborhood_room_type_avg_price['neighbourhood'].isin(top_5_expensive_neighborhoods['neighbourhood'])]

# Sort neighborhoods by average price per night and select top 5 cheapest
top_5_cheapest_neighborhoods = neighborhood_avg_price.sort_values(by='price_per_night').head(5)

# Find the room types in these top 5 neighborhoods
top_5_cheapest_details = neighborhood_room_type_avg_price[neighborhood_room_type_avg_price['neighbourhood'].isin(top_5_cheapest_neighborhoods['neighbourhood'])]

print("Top 5 Most Expensive Neighborhoods with Room Types:")
print(top_5_expensive_details)

print("\nTop 5 Cheapest Neighborhoods with Room Types:")
print(top_5_cheapest_details)

# Set the plot style
sns.set(style="whitegrid")

# Define custom color palettes for room types
unique_room_types = dfClean['room_type'].unique()
colors_expensive = sns.color_palette("husl", len(unique_room_types))
colors_cheapest = sns.color_palette("husl", len(unique_room_types))

# Plot for Top 5 Most Expensive Neighborhoods
plt.figure(figsize=(18, 8))
plt.subplot(1, 2, 1)
sns.barplot(x='price_per_night', y='neighbourhood', hue='room_type', data=top_5_expensive_details, palette=dict(zip(unique_room_types, colors_expensive)))
plt.title('Top 5 Most Expensive Neighborhoods and Room Types', fontsize=16)
plt.xlabel('Average Price per Night', fontsize=14)
plt.ylabel('Neighborhood', fontsize=14)
plt.legend(title='Room Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot for Top 5 Cheapest Neighborhoods
plt.subplot(1, 2, 2)
sns.barplot(x='price_per_night', y='neighbourhood', hue='room_type', data=top_5_cheapest_details, palette=dict(zip(unique_room_types, colors_cheapest)))
plt.title('Top 5 Cheapest Neighborhoods and Room Types', fontsize=16)
plt.xlabel('Average Price per Night', fontsize=14)
plt.ylabel('Neighborhood', fontsize=14)
plt.legend(title='Room Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
# Hitung harga rata-rata sewa untuk setiap wilayah dan tipe ruangan
average_price_by_neighbourhood_and_room_type = dfClean.groupby(['neighbourhood', 'room_type'])['price'].mean().reset_index()

# Tampilkan 5 wilayah dengan harga sewa rata-rata tertinggi untuk setiap tipe ruangan
room_types = dfClean['room_type'].unique()
top_5_expensive = {}
top_5_cheap = {}

for room_type in room_types:
    filtered_data = average_price_by_neighbourhood_and_room_type[average_price_by_neighbourhood_and_room_type['room_type'] == room_type]
    top_5_expensive[room_type] = filtered_data.nlargest(5, 'price')
    top_5_cheap[room_type] = filtered_data.nsmallest(5, 'price')

# Menampilkan hasil
print("5 Wilayah dengan Harga Sewa Rata-rata Tertinggi untuk Setiap Tipe Ruangan:")
for room_type, dfClean in top_5_expensive.items():
    print(f"\nTipe Ruangan: {room_type}")
    print(dfClean)

print("\n5 Wilayah dengan Harga Sewa Rata-rata Terendah untuk Setiap Tipe Ruangan:")
for room_type, dfClean in top_5_cheap.items():
    print(f"\nTipe Ruangan: {room_type}")
    print(dfClean)
```
